## Collecting Multiplier, Character, Power Level and Saga or Movie Data using Web Scrapping

```python
from bs4 import BeautifulSoup
import urllib.request,urllib.error,urllib.response
import re
url=urllib.request.urlopen("https://www.deviantart.com/mimozukakito-kun/journal/All-Dragon-Ball-Power-Levels-373101310").read()
soup=BeautifulSoup(url,"html.parser")
tags=soup("li")
Multipliers=list()
C_Pl_Data=list()
for tag in tags:
    mul_data=re.findall("<li><b>(.+?)</b></li>",str(tag))
    if(len(mul_data)>0):
        for md in mul_data:
            Multipliers.append(md)
    cpl=re.findall("<li>(.+)</li>",str(tag))
    if(len(cpl)>0):
        for cpd in cpl:
            C_Pl_Data.append(cpd.split(": "))
tags=soup("h2")
temp_Saga_or_Movie=list()
for tag in tags:
    sm=re.findall("<h2>(.+)</h2>",str(tag))
    if(len(sm)>0):
        temp_Saga_or_Movie.append(sm)
```

## Cleaning the Collected Data

### Slicing the Character and Power Level Data aone

```python
C_Pl_Data=C_Pl_Data[13:]
C_Pl_Data=C_Pl_Data[:-1]
```

### Cleaning the Saga or Movie list

```python
for data in temp_Saga_or_Movie:
    for value in data:
        print(value)
        value=value.replace("<b>","")
        value=value.replace("</b>","")
        value=value.replace("\xa0","")
        if(value=="Dragon Ball" or value=="Dragon Ball Z" or value=="Dragon Ball GT" or value=="Movies and Specials"):
            temp_Saga_or_Movie.remove(data)
        print(value)
```
